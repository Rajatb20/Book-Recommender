# Popularity based recommender 

```python
# Merging two datas to get books with ratings in one dataset

ratings_of_books = ratings.merge(books, on = 'ISBN')
ratings_of_books.head()
```

```python
# To find number of ratings given to each book so that we can find the top 50 books with highest ratings

num_of_ratings = ratings_of_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_of_ratings.rename(columns={'Book-Rating':'Num of Ratings'},inplace = True)
num_of_ratings
```

```python
# # To find avarage ratings given to each book so that we can find the books with highest avarage ratings

avg_ratings = ratings_of_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating':'avg of Ratings'},inplace = True)
avg_ratings
```

```python
# To find the  top 50 books which have more than 250 ratings

popular_books = num_of_ratings.merge(avg_ratings, on = 'Book-Title')
popular_books = popular_books[popular_books['Num of Ratings'] >= 250].sort_values("avg of Ratings",ascending = False).head(50)
popular_books
```

```python
# Merging the two datas to get the required columns

popular = popular_books.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Publisher','Num of Ratings','avg of Ratings','Image-URL-M']]
```

# Collaborative Filtering Based Recommender System

```python
ratings_of_books.head()
```

```python
x = ratings_of_books.groupby('User-ID').count()['Book-Rating'] > 200
users_with_high_ratings = x[x].index
```

```python
filtered_ratings = ratings_of_books[ratings_of_books['User-ID'].isin(users_with_high_ratings)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] > 50
books_with_high_ratings = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(books_with_high_ratings)]
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse = True)[1:5]
    
    data = []
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```
